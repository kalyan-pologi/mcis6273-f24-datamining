## Load the dataset

```python
df = pd.read_csv("./hw0_dataset_1M.csv", names=[i for i in range(30)])
```

```python
# Identify rows that potentially contain header information
header_candidates = df[df[1].str.contains('key', na=False)]
header_candidates
```

```python
# Convert all elements to string, then find the longest row by concatenating its string elements
headers = header_candidates.loc[header_candidates.apply(lambda x: len(' '.join(x.astype(str).dropna())), axis=1).idxmax()]
```

```python
# Set new headers
df.columns = headers.values
```

```python
# Drop all header candidate rows from the DataFrame to clean it
df = df.drop(header_candidates.index)
```

```python
df.head()
```

```python
# df.info()
```

```python
df = df.iloc[ : , : 10]
df.head()
```

```python
df.dropna(inplace = True)
```

```python
# Save the cleaned DataFrame
# df.to_csv('/cleaned_data.csv', index=False)
```

```python
# Restricting columns to 'DOI', 'journal-title', 'doi-asserted-by'
df = df.iloc[:, [3,2,9]] 
df.head()
```

```python
# Rename columns
df.columns = ['DOI', 'doi-asserted-by', 'journal-title']  
```

```python
# Resetting the index of the DataFrame
df = df.reset_index(drop=True)
df.head()
```

```python
# Normalize the 'DOI' column: convert to lowercase
df['DOI'] = df['DOI'].str.lower()
```

```python
df
```

```python
# Normalize the 'journal-title' column: 
# Remove punctuation
df['journal-title'] = df['journal-title'].str.replace('[^\w\s]', '', regex=True) 
```

```python
df
```

```python
# Remove extra spaces
df['journal-title'] = df['journal-title'].str.replace('\s+', ' ', regex=True).str.strip()  
```

```python
df
```

```python
# Export the DataFrame to CSV and JSON
# df.to_csv('/data_cleaned_normalized.csv', index=False)
# df.to_json('/data_cleaned_normalized.json', orient='records')
```

```python
# Filter the DataFrame to only include rows where both 'DOI' and 'journal-title' are not null
filtered_df = df.dropna(subset=['DOI', 'journal-title'])
```

```python
# Remove duplicates based on the 'DOI' column to ensure each DOI is unique
filtered_df = filtered_df.drop_duplicates(subset='DOI')
```

```python
# Returns a DataFrame containing the n most frequent DOIs
def n_most_frequent_doi(df, n):
    # Count the occurrences of each DOI
    freq = df['DOI'].value_counts().head(n)
    # Convert the Series to a DataFrame
    return freq.reset_index().rename(columns={'index': 'DOI', 'DOI': 'Count'})
```

```python
# Returns a DataFrame containing rows where the journal title contains a specific substring
def journal_lookup(df, s):
    # Filter the DataFrame based on substring presence in 'journal-title'
    return df[df['journal-title'].str.contains(s, na=False, case=False)]
```
